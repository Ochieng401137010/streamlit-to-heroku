# Data Exploratory Analysis

```python
# Plot the dataset
color_list = [
    'blue',
    'orange',
    'green',
    'red',
    'purple',
    'brown',
    'pink',
    'gray',
    'olive',
    'cyan',
]

def Visualize(data):
    features = list(df.select_dtypes(include=[np.number]).columns.values)
    feature_size = len(features)
    fig, axes = plt.subplots(nrows=int(np.ceil(feature_size/2)), ncols=2, figsize=(14, feature_size*2), dpi=80, 
                             facecolor='w', edgecolor='k')
    
    for i in range(feature_size):
        key = features[i]
        c = color_list[i % (len(color_list))]
        t_data = data[key]
        t_data.head()
        ax = t_data.plot(
            ax=axes[i // 2, i % 2],
            color=c,
            title='{}'.format(key),
            rot=25),
        
        plt.tight_layout()
        
Visualize(df)
```

From the observation of the above plots,it shows that the car attributes are stationary as they are not dependent on time. They do not have specific trend or seasonal effects hence they can easily be modelled.

# Analyzing Data

Create different plot to check relationship between variables, that is; how one variable is affecting the other.

```python
sns.countplot(x="fueltype", data=df)
```

From the above plot, most cars are using gas as the type of fuel as compared to diesel.

```python
sns.countplot(x="enginetype", data=df)
```

From the observation, we can conclude that ohc is the most commonly used car engine as it normally has the advantage of greater flexibilty to optimize the size, location and shape of the intake and exhaust ports, since there are no pushrods to be avoided.

```python
sns.countplot(x="carbody", data=df)
```

From the analysis, the most frequent type of carbody are hatchback and sedan carbody.

```python
df["CarName"].value_counts().head(10).plot(kind="bar")
```

From the above plots, toyota corona, peugeot 504 and toyota corolla are the most used cars.

```python
df["horsepower"].plot.hist(figsize=(10,5))
```

The above histogram shows that most cars has 50 to 125 horsepower. This implies that most of these cars has a low acceleration rate due to the low horsepower production.

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
a = df.groupby('aspiration').aspiration.count()
b = df.aspiration.unique()
ax.pie(a, labels=b, autopct='%1.1f%%')
plt.show()
```

From the above piechart, most cars uses the standard aspiration engine whereas only 18% of the car uses the turbo aspiration engine. This also reconfirms our above analysis that these cars could have a low acceleration rate.

```python
alt.Chart(df).mark_circle().encode(
    x = "CarName",
    y = "price",
    color = 'citympg',
    tooltip = ["compressionratio","wheelbase","car_ID","CarName"]
)
```

```python
alt.Chart(df).mark_circle().encode(
    x = "price",
    y = "horsepower",
    color = 'citympg',
    tooltip = ["compressionratio","wheelbase","car_ID","CarName"]
)
```

The above altair graph explore the possible relation between the price and hoursepower of a vehicle. Additional infomation about the selected vehicle are display.
